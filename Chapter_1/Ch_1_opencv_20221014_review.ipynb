```python
import cv2
import numpy as np
import sys
```

```python
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR) # default 는 cv2.IMREAD_COLOR

# 사이즈 변경
img_resize = cv2.resize(img, (100,100), interpolation = cv2.INTER_LINEAR)
cv2.imwrite('img_resize.png', img_resize)

if img is None :
    print('Image read failed')
    sys.exit()
    
# window 창을 생성 -> 이름음 image
cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE) # default 는 cv2.WINDOW_AUTOSIZE

# 만든 window의 이름을 넣어준다.
cv2.imshow('image', img) # 영상을 띄어준다.
cv2.imshow('image_resize', img_resize)


while True :
    key = cv2.waitKey()
    if key == 27 or key == ord('q') : # key에 esc 나 q 가 들어오면 종료
        break

cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import matplotlib.pyplot as plt

img = cv2.imread('./fig/cat.bmp')
# RGG 로 변경
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

if img is None :
    print('image read failed')
    sys.exit()
    
cv2.imshow('cat', img)

plt.imshow(imgRGB)
plt.axis('off')
plt.show()

while True :
    key = cv2.waitKey()
    
    if key == 27 or key == ord('q') :
        break
        
cv2.destroyAllWindows()
```

### 폴더 내 파일 확인

```python
os.listdir('./fig/images/')
```

```python
img_lists = os.listdir('./fig/images/')

img_files = []
for i in img_lists :
    img_file = './fig/images/' + i
    img_files.append(img_file)
    
print(img_files)

# 윈도우 생성
cv2.namedWindow('scenery', cv2.WINDOW_NORMAL)

# cv2.setWindowProperty('scenery', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

# img_files 에 존재하는 폴더를 읽어 온다.
for i in img_files :
    img = cv2.imread(i)
#     print(i)
    if img is None :
        print('image read failed')
        sys.exit()
    
    cv2.imshow('scenery', img)
    
    if cv2.waitKey(3000) == 27 :
        break
        
cv2.destroyAllWindows()
cv2.waitKey(1)
```

## 지속적으로 읽을때

```python
img_lists = os.listdir('./fig/images/')

img_files = []
for i in img_lists :
    img_file = './fig/images/' + i
    img_files.append(img_file)
    
print(img_files)

# 윈도우 생성
cv2.namedWindow('scenery', cv2.WINDOW_NORMAL)

# cv2.setWindowProperty('scenery', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

idx = 0

while True :
    img = cv2.imread(img_files[idx%len(img_files)])
    
    if img is None :
        print('image read failed')
        break
    
    cv2.imshow('scenery', img)
    
    if cv2.waitKey(3000) == 27 :
        break
    
    idx += 1
        
cv2.destroyAllWindows()
cv2.waitKey(1)
```
